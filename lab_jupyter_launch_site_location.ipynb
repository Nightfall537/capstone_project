```python
# Task 2: Mark the success/failed launches for each site on the map
import folium
import pandas as pd
from folium.plugins import MarkerCluster

# Assuming you have a DataFrame named 'launch_sites_df' with columns:
# 'Launch Site', 'Lat', 'Long', and 'Class' (where 1 indicates success, 0 failure)
# Example DataFrame (replace with your actual data)
data = {'Launch Site': ['CCSFS SLC 40', 'VAFB SLC 4E', 'KSC LC 39A', 'CCSFS SLC 40', 'VAFB SLC 4E'],
        'Lat': [28.561857, 34.632093, 28.573255, 28.561857, 34.632093],
        'Long': [-80.577366, -120.610829, -80.646895, -80.577366, -120.610829],
        'Class': [1, 0, 1, 0, 1]}  # 1: success, 0: failure
launch_sites_df = pd.DataFrame(data)

# Create a Folium map centered on the first launch site
map_center = [launch_sites_df['Lat'].iloc[0], launch_sites_df['Long'].iloc[0]]
launch_map = folium.Map(location=map_center, zoom_start=4)

# Create Marker Clusters for better map performance with many points
marker_cluster = MarkerCluster().add_to(launch_map)

# Loop through the launch sites and add markers with color based on success/failure
for index, row in launch_sites_df.iterrows():
    coordinate = [row['Lat'], row['Long']]
    launch_site_name = row['Launch Site']
    launch_class = row['Class']

    if launch_class == 1:
        color = 'green'  # Success
        popup_text = f"{launch_site_name} - Success"
    else:
        color = 'red'  # Failure
        popup_text = f"{launch_site_name} - Failure"

    folium.Marker(
        location=coordinate,
        popup=popup_text,
        icon=folium.Icon(color=color)
    ).add_to(marker_cluster)  # Add to the MarkerCluster

# Display the map

launch_map 
```

```python
# TASK 3: Calculate the distances between a launch site to its proximities
import pandas as pd
from math import sin, cos, sqrt, atan2, radians

def calculate_distance(lat1, lon1, lat2, lon2):
    """
    Calculates the distance between two points on the Earth (latitude, longitude).
    """
    # approximate radius of Earth in km
    R = 6373.0

    lat1 = radians(lat1)
    lon1 = radians(lon1)
    lat2 = radians(lat2)
    lon2 = radians(lon2)

    dlon = lon2 - lon1
    dlat = lat2 - lat1

    a = sin(dlat / 2)**2 + cos(lat1) * cos(lat2) * sin(dlon / 2)**2
    c = 2 * atan2(sqrt(a), sqrt(1 - a))

    distance = R * c
    return distance

# Example DataFrame (replace with your actual data)
data = {'Launch Site': ['CCSFS SLC 40', 'VAFB SLC 4E', 'KSC LC 39A'],
        'Lat': [28.561857, 34.632093, 28.573255],
        'Long': [-80.577366, -120.610829, -80.646895],
        'Proximity_Lat': [28.5215, 34.6432, 28.5750], # Example Proximity Latitudes.
        'Proximity_Long': [-80.6475, -120.5760, -80.6040]} # Example Proximity Longitudes.

launch_sites_df = pd.DataFrame(data)

# Calculate distances
launch_sites_df['Distance_to_Proximity'] = launch_sites_df.apply(
    lambda row: calculate_distance(row['Lat'], row['Long'], row['Proximity_Lat'], row['Proximity_Long']), axis=1
)

print(launch_sites_df)
```

```python
# Create a marker with distance to a closest city, railway, highway, etc.
# Draw a line between the marker to the launch site

import folium
import pandas as pd


# Assuming you have a DataFrame named 'launch_sites_df' with columns 'Launch Site', 'Lat', 'Long'
# Example DataFrame (replace with your actual data)
data = {'Launch Site': ['CCSFS SLC 40', 'VAFB SLC 4E', 'KSC LC 39A'],
        'Lat': [28.561857, 34.632093, 28.573255],
        'Long': [-80.577366, -120.610829, -80.646895]}
launch_sites_df = pd.DataFrame(data)

# Closest city coordinates (replace with your actual data)
city_coords = {
    'CCSFS SLC 40': {'Lat': 28.3963, 'Long': -80.6048, 'Name': 'Titusville'},
    'VAFB SLC 4E': {'Lat': 34.6953, 'Long': -120.4579, 'Name': 'Lompoc'},
    'KSC LC 39A': {'Lat': 28.6139, 'Long': -80.8075, 'Name': 'Orlando'}
}

def calculate_distance(lat1, lon1, lat2, lon2):
    """Calculates the distance between two points on the Earth."""
    from math import sin, cos, sqrt, atan2, radians
    R = 6373.0
    lat1 = radians(lat1)
    lon1 = radians(lon1)
    lat2 = radians(lat2)
    lon2 = radians(lon2)
    dlon = lon2 - lon1
    dlat = lat2 - lat1
    a = sin(dlat / 2)**2 + cos(lat1) * cos(lat2) * sin(dlon / 2)**2
    c = 2 * atan2(sqrt(a), sqrt(1 - a))
    distance = R * c
    return distance

# Create a Folium map centered on the first launch site
map_center = [launch_sites_df['Lat'].iloc[0], launch_sites_df['Long'].iloc[0]]
launch_map = folium.Map(location=map_center, zoom_start=6)

# Add launch site markers
for index, row in launch_sites_df.iterrows():
    folium.Marker(
        location=[row['Lat'], row['Long']],
        popup=row['Launch Site'],
        icon=folium.Icon(color='blue')
    ).add_to(launch_map)

# Add city markers, distance labels, and lines to launch sites
for index, row in launch_sites_df.iterrows():
    launch_site_name = row['Launch Site']
    launch_site_lat = row['Lat']
    launch_site_lon = row['Long']

    city_lat = city_coords[launch_site_name]['Lat']
    city_lon = city_coords[launch_site_name]['Long']
    city_name = city_coords[launch_site_name]['Name']

    distance_city = calculate_distance(launch_site_lat, launch_site_lon, city_lat, city_lon)

    # Add city marker with distance label
    folium.Marker(
        location=[city_lat, city_lon],
        popup=f"Closest City: {city_name}",
        icon=DivIcon(
            icon_size=(100, 20),
            icon_anchor=(0, 0),
            html=f'<div style="font-size: 12; color:#d35400;"><b>{distance_city:10.2f} KM</b></div>',
        )
    ).add_to(launch_map)

    # Draw a line between the launch site and the city
    folium.PolyLine(
        locations=[[launch_site_lat, launch_site_lon], [city_lat, city_lon]],
        color='red',
        weight=2.5,
        opacity=1
    ).add_to(launch_map)

# Display the map

launch_map
```
